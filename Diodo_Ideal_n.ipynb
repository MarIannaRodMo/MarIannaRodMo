# Ecuación del diodo Ideal

## Objetivo:

 - Graficar la ecuación del diodo ideal

 $$ I = I_o[\exp{(\frac{V_D}{V_T})} -1 ]$$

 Donde:

 - $V_T = \frac{nK_BT}{q}$
 - $K_B = 1.38x10^{-23} \text{J/K}$
 - $T = 300 \, \text{K}$
 - $n = 1 $

 ##### de tal forma que:

  $$ I = I_o[\exp{(\frac{q V_D}{n K_B T})} -1 ]$$

  ##### La constante de idealidas en factor que varía según el dispositivo y material. A continuación, se muestran algunas gráficas del diodo con distintos valores para dicha constante.

```python
## Importar librerias

import numpy as np
import matplotlib.pyplot as plt
```

```python
## Declarar constantes

Io = 1e-6    #[A] Corriente de saturación
n = 1         # Factor de idelidad
Kb =1.38e-23  # [J/K] Constante de boltzman
q =1.6e-19   # [C] Carga del electrón
T = 300       # [K] Temperatura ambiente
```

```python
## Voltaje termico

V_T = (Kb*T)/q
round(V_T,3)
```

```python
## Variación de voltaje  (eje-x)

V = np.linspace(-1, 1, num=200)
```

```python
# Ecuación del diodo

I = Io*(np.exp(V/(n*V_T))-1)/1e-3 # [mA]
```

```python
## Graficar

fig,axes=plt.subplots()
axes.plot(V,I)
axes.set_ylim(-1.9, 45)
##axes.fig.add_axes(0)
##plt.axline        esto creo que me va a servir para proyectarla


axes.set_xlabel("Voltaje del Diodo [V]")
axes.set_ylabel("Corriente [mA]")
axes.set_title("Curva I-V del Diodo rectificador N00")
plt.show()
```
